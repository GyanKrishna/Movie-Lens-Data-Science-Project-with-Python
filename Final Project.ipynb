```python
Genres = Movies['Genres']
```

```python
x=[]
for each in Genres.unique():
    x.extend(each.split('|'))  
```

```python
Genres_Value=pd.Series(x)
```

```python
Genres_Unique_Val=Genres_Value.unique()
```

```python
Genres_Unique_Val
```

```python
#Create a separate column for each genre category with a one-hot encoding ( 1 and 0) whether or not the movie belongs to that genre. 
```

```python
for each in Genres_Unique_Val:
    Ser=Movies['Genres'].apply(lambda x: each in x)
    if each == 'Animation':
            Animation = np.array(Ser)
    elif each =="Children's":
            Children_s= np.array(Ser)
    elif each =='Comedy':
            Comedy= np.array(Ser)
    elif each =='Adventure':
            Adventure= np.array(Ser)
    elif each =='Fantasy':
            Fantasy= np.array(Ser)
    elif each =='Romance':
            Romance= np.array(Ser)
    elif each =='Drama':
            Drama= np.array(Ser)
    elif each =='Action':
            Action= np.array(Ser)
    elif each =='Crime':
            Crime= np.array(Ser)
    elif each == 'Thriller':
            Thriller = np.array(Ser)
    elif each =='Horror':
            Horror= np.array(Ser)
    elif each =='Sci-Fi':
            Sci_Fi= np.array(Ser)
    elif each =='Documentary':
            Documentary= np.array(Ser)
    elif each =='War':
            War= np.array(Ser)
    elif each =='Musical':
            Musical= np.array(Ser)
    elif each =='Mystery':
            Mystery= np.array(Ser)
    elif each =='Film-Noir':
            Film_Noir= np.array(Ser)
    elif each =='Western':
            Western = np.array(Ser)
```
